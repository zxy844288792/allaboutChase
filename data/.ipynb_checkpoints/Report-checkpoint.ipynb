```python
import matplotlib
import matplotlib.ticker as plticker
import matplotlib.pyplot as plt
%matplotlib inline

def parse_data(filename, tag, data_idx):
    data = []
    with open(filename, "r") as f:
        for line in f:
            words = line.split(" ")
            if (words[0] == tag):
                data.append(float(words[data_idx + 1]))
    return data

def plot_lines(ys, labels, title, x_axis, y_axis, spacing=30.48):
    fig, ax = plt.subplots()
    x = range(len(ys[0]))
    plt.xlabel(x_axis)
    plt.ylabel(y_axis)
    
    #spacing = 2.54
    #minorLocator = plticker.MultipleLocator(spacing)
    #ax.yaxis.set_minor_locator(minorLocator)
    #ax.grid(which='minor')
    majorLocator = plticker.MultipleLocator(spacing)
    ax.yaxis.set_major_locator(majorLocator)
    ax.grid(which='major')
    for i, y in enumerate(ys):
        plt.plot(x,y, label=labels[i])
    plt.legend()
    plt.title(title)
    plt.savefig("%s.png"%title)
```

```python
xs = parse_data('log_z.txt', "p:", 0)
ys = parse_data('log_z.txt', "p:", 1)
zs = parse_data('log_z.txt', "p:", 2)

plot_lines([xs, ys, zs], ['x', 'y', 'z'], 'z-distance', 'number of frames', 'distance in cm')
```

```python
xs = parse_data('log_x.txt', "p:", 0)
ys = parse_data('log_x.txt', "p:", 1)
zs = parse_data('log_x.txt', "p:", 2)

plot_lines([xs, ys, zs], ['x', 'y', 'z'], 'x-distance', 'number of frames', 'distance in cm', 2.45*3)
```

```python
u0 = parse_data('log_large_u.txt', "u:", 1)

processed_u0=[]
processed_u0.append(0)
for u in u0:
    if u > 400:
        processed_u0.append(400)
    else:
        processed_u0.append(u)
plot_lines([processed_u0], ['distance'], 'ultrasonic-distance-large-object', 'number of frames', 'distance in cm', 30.43)
```
